```python
#import dataset
import pandas as pd
df=pd.read_csv("marketing_campaign.csv", sep='\t')
df.head()
```

```python
df.info()
```

```python
#check from null columns 
df.isnull().sum()
```

```python
#check fro duplicate data
df.duplicated().sum()
```

```python
print("before cleaning")
print(df['Education'].unique())
df['Education'] = (df['Education']
                   .str.strip()
                   .str.lower()
                   .str.title()
                  )
print("after cleaning")
print(df['Education'].unique())
```

```python
print("before cleaning")
print(df['Marital_Status'].unique())
df['Marital_Status'] = (df['Marital_Status']
                   .str.strip()
                   .str.lower()
                   .str.title()
                  )
print("after cleaning")
print(df['Marital_Status'].unique())
```

```python
df[df.isnull().any(axis=1)]
```

```python
#converting null values into 0
df['Income']= df['Income'].fillna(0)
#checking for null values
df[df.isnull().any(axis=1)]
```

```python
df['Income'].head(11)
```

```python
#converting Dt_Customer datatype object to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], 
                                   format='%d-%m-%Y', 
                                   errors='coerce') 
```

```python
#check datatype of Dt_Customer
df.info()
```

```python
#Rename column headers to be clean and uniform (e.g., lowercase, no spaces).
df.columns = df.columns.str.strip().str.lower()
```
